```python
"""
  Week 8 Lab
  Level: 1
  Name: Alexis Hollingsworth
  UID: U1275318
"""

import os

# Function to display the menu
def display_menu():
    print("\nExpense Tracker Menu:")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Edit Expense")
    print("4. Delete Expense")
    print("5. Exit")

# Function to add an expense
def add_expense():
    date = input("Enter date (YYYY-MM-DD): ")
    category = input("Enter category: ")
    description = input("Enter description: ")
    amount = input("Enter amount: ")

    # Check if amount is a valid number
    try:
        amount = float(amount)
    except ValueError:
        print("Invalid amount. Please enter a numeric value.")
        return

    # Append the expense to the file
    with open("expenses.csv", "a") as file:
        file.write(f"{date},{category},{description},{amount}\n")

    print("Expense added successfully!")

# Function to view all expenses
def view_expenses():
    if not os.path.exists("expenses.csv"):
        print("No expenses recorded.")
        return

    with open("expenses.csv", "r") as file:
        expenses = file.readlines()

    if not expenses:
        print("No expenses recorded.")
    else:
        print("\nRecorded Expenses: ")
        for idx, expense in enumerate(expenses, start=1):
            date, category, description, amount = expense.strip().split(",")
            print(f"{idx}. Date: {date}, Category: {category}, Description: {description}, Amount: ${amount}")

# Function to edit an expense
def edit_expense():
    view_expenses()

    try:
        expense_number = int(input("\nEnter the number of the expense you want to edit: "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return

    with open("expenses.csv", "r") as file:
        expenses = file.readlines()

    if expense_number < 1 or expense_number > len(expenses):
        print("Invalid expense number.")
        return

    expense = expenses[expense_number - 1].strip().split(",")
    date, category, description, amount = expense

    print(f"\nEditing Expense: {date}, {category}, {description}, ${amount}")
    new_date = input(f"Enter new date (press enter to keep '{date}'): ") or date
    new_category = input(f"Enter new category (press enter to keep '{category}'): ") or category
    new_description = input(f"Enter new description (press enter to keep '{description}'): ") or description
    new_amount = input(f"Enter new amount (press enter to keep '{amount}'): ") or amount

    # Check if new amount is a valid number
    try:
        new_amount = float(new_amount)
    except ValueError:
        print("Invalid amount. Please enter a numeric value.")
        return

    # Update the expense in the list
    expenses[expense_number - 1] = f"{new_date},{new_category},{new_description},{new_amount}\n"

    # Save the updated expenses back to the file
    with open("expenses.csv", "w") as file:
        file.writelines(expenses)

    print("Expense updated successfully!")

# Function to delete an expense
def delete_expense():
    view_expenses()

    try:
        expense_number = int(input("\nEnter the number of the expense you want to delete: "))
    except ValueError:
        print("Invalid input. Please enter a valid number.")
        return

    with open("expenses.csv", "r") as file:
        expenses = file.readlines()

    if expense_number < 1 or expense_number > len(expenses):
        print("Invalid expense number.")
        return

    # Remove the selected expense
    expenses.pop(expense_number - 1)

    # Save the updated expenses back to the file
    with open("expenses.csv", "w") as file:
        file.writelines(expenses)

    print("Expense deleted successfully.")

# Main function to run the menu loop
def expense_tracker():
    while True:
        display_menu()
        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == '1':
            add_expense()
        elif choice == '2':
            view_expenses()
        elif choice == '3':
            edit_expense()
        elif choice == '4':
            delete_expense()
        elif choice == '5':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the program
if __name__ == "__main__":
    expense_tracker()
```
